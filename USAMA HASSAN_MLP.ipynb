```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Activation
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
batch_size = 128
LEARNING_RATE = 1e-2
num_classes = 2
epochs = 5000

features = datasets.load_breast_cancer().data

    # standardize the features
features = StandardScaler().fit_transform(features)
print(features.shape)
    # get the number of features
num_features = features.shape[1]
print(num_features)
    # load the labels for the features
labels = datasets.load_breast_cancer().target

train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size=0.30)
model = Sequential()
model.add(Dense(30,input_dim=num_features))
model.add(Activation('relu'))
model.add(Dense(1,activation='sigmoid'))


model.summary()
```

```python

model.compile(loss=keras.losses.binary_crossentropy,optimizer=keras.optimizers.SGD(lr=0.1, momentum=0.9, nesterov=True),metrics=['accuracy'])

model.fit(train_features, train_labels , batch_size = batch_size, epochs = epochs, verbose=1, validation_data=(test_features, test_labels))
```
